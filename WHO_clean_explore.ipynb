### Dealing with outliers and damaged data points

The most obvious dirt that is left to be cleaned in this dataset is that there are a huge amount of measurement errors, where a percentage value is 200 or a value that is obviously 1 or 2 decimal places off from the real value. 

```python
full_df.tail()
```

To start I will be dropping several highly damaged columns or ones that aren't relevant to our life expectancy analysis. Adult mortality isn't worth keeping, bc it's almost the same measurement as life expectancy. I'll drop percentage expenditure and measles because they are obviously damaged areas of the dataset with many non-sensical values throughout the column.

```python
##drop adult mortality and percentage expenditure
full_df = full_df.drop(columns=['Adult Mortality', 'percentage expenditure'])
```

```python
full_df.head(20)
```

We still have some remaining columns with outliers due to measurement errors. In order to efficiently deal with these bad values, I'm going to winsorize them to remove any egregious errors.

```python
from scipy.stats.mstats import winsorize
winsor_df = full_df
```

```python
##list of which columns not to winsorize
ignore = ['Country', 'Year', 'Status']
##loop through columns and winsorize them
for column in list(winsor_df):
    if column not in ignore:
        winsor_df[column] = winsorize(winsor_df[column], (.1,.1), inplace=True)
```

```python
winsor_df.head(100)
```

Now that they are all winsorized, there's only one problem left to deal with. GDP numbers for some countries are obviously incorrerct, where a certain very low or very high value was brought closer to the true value by winsorization, but is still way off. Unlike something like a measles outbreak, GDP values will not deviate so vastly, even in a financial crisis. To solve this, I'm going to replace each GDP value with that country's median GDP value, in order to cut out such innaccurate outliers.

```python
winsor_df['GDP'] = winsor_df.groupby('Country')['GDP'].transform(lambda x: x.median())
```

```python
winsor_df.head(100)
```
